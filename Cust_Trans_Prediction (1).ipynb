```python
print("y_true:", y_true)
```

```python
confusion_matrix = np.array([[44306, 574],
                            [3937, 1183]])
```

```python
print("y_true:", y_true)
```

```python
y_true = []
y_pred = []
```

```python
for true_label in range(confusion_matrix.shape[0]):
    for pred_label in range(confusion_matrix.shape[1]):
        count = confusion_matrix[true_label, pred_label]
        y_true.extend([true_label] * count)
        y_pred.extend([pred_label] * count)
```

```python
#Convert lists to numpy arrays
y_true = np.array(y_true)
y_pred = np.array(y_pred)
```

```python
print("y_true:", y_true)
print("y_pred:", y_pred)
```
